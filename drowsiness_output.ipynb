```python
import ipywidgets as widgets
from IPython.display import display, Image
from PIL import Image as PILImage
import textwrap

# Create a file uploader widget
uploader = widgets.FileUpload(description='Upload Image', accept='.jpg,.jpeg,.png')

# Create a button widget
button = widgets.Button(description='Process')

# Create an output widget for displaying the result
output = widgets.Output()

n = 2

# Define the function to process the uploaded image
def process_image(change):
    
    # Clear the output widget
    output.clear_output()

    # Get the uploaded image
    image = uploader.data[0]
    
    # Save the image to a file
    image_path = 'uploaded_image.jpg'
    with open(image_path, 'wb') as f:
        f.write(image)
    
    # Run the face detection model on the image
    !python /content/yolov5/detect.py --weights /content/drive/MyDrive/dataipd/best.pt --source {image_path} --save-txt
    label_file = f'/content/yolov5/runs/detect/exp{n}/labels/uploaded_image.txt'
    
    # Read the content of the label.txt file
    with open(label_file, 'r') as file:
        line = file.readline().strip()
    
    # Extract the bounding box coordinates
    values = line.split()
    bbox = [float(value) for value in values[1:]]

    # Open the image
    image = PILImage.open(image_path)

    # Get the image dimensions
    width, height = image.size

    # Calculate the coordinates of the bounding box
    left = int((bbox[0] - bbox[2] / 2) * width)
    top = int((bbox[1] - bbox[3] / 2) * height)
    right = int((bbox[0] + bbox[2] / 2) * width)
    bottom = int((bbox[1] + bbox[3] / 2) * height)

    # Crop the image using the calculated coordinates
    cropped_image = image.crop((left, top, right, bottom))
    
    # Save the cropped image to a file
    cropped_image_path = 'cropped_image.jpg'
    cropped_image.save(cropped_image_path)
    
    # Run the eye and yawn detection model on the cropped image
    !python /content/yolov5/classify/predict.py --weights /content/content/yolov5/runs/train-cls/exp/weights/best.pt --source {cropped_image_path}
    !python /content/yolov5/classify/predict.py --weights /content/content/yolov5/runs/train-cls/exp2/weights/best.pt --source {image_path}
    final_image_path = f'/content/yolov5/runs/detect/exp{n}/uploaded_image.jpg'

    with output:
        print("")
        print("INPUT IMAGE")
        print("")
        display(Image(filename=image_path))
        print("")
        print("")
        print("OUTPUT IMAGE")
        with open(f"/content/yolov5/runs/predict-cls/exp{n}/output.txt", 'r') as file:
          content = file.read()
          wrapped_text = textwrap.fill(content, width=100)
          print(wrapped_text)
        with open(f"/content/yolov5/runs/predict-cls/exp{n+1}/output.txt", 'r') as file:
          content = file.read()
          wrapped_text = textwrap.fill(content, width=100)
          print(wrapped_text)
        display(Image(filename=final_image_path))
        print("")
        print("")

# Register the process_image function to be called when the button is clicked
button.on_click(process_image)

# Display the UI
display(uploader, button, output)
```
